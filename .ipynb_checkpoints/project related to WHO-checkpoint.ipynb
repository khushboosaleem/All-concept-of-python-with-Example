```python
def plot_cases(country_name):
    bol = who["Country"] == country_name
    df = who[bol]
    plt.plot(df["Date_reported"],df["Cumulative_cases"])
    plt.xlabel("Date_reported")
    plt.ylabel("Cumulative_cases")
    plt.title("{} : Covide cummulative cases".format(country_name))     
    plt.show()
    
```

```python
plot_cases("India")  ## exponential grapgh esa graph jo ahisa ahista bhary or phir ak dam sy bar jay
```

```python
plot_cases("China")
```

```python
plot_cases("Grenada")
```

```python
plot_cases("Comoros")
```

```python
# The graph is increasing constantly that is called linear growth 
plot_cases("Poland")
```

```python
bol1 = who["Country"] == "The United Kingdom"
united_kingdom = who[bol1]
united_kingdom
```

```python
bol2 = who["Country"] == "France"
France = who[bol2]
France
```

```python
plt.plot(united_kingdom["Date_reported"], united_kingdom["Cumulative_cases"], label = "UK")
plt.plot(France["Date_reported"], France["Cumulative_cases"] , label = "France")
plt.title("UK and France Covid Cases")
plt.xlabel("Date reported")
plt.ylabel("Cumulative casses")
plt.legend()
plt.show()
```

```python
bol3 = who["Country"] == "Pakistan"
pakistan = who[bol3]
pakistan
```

```python
plt.plot(pakistan["Date_reported"], pakistan["Cumulative_deaths"])
plt.xlabel("Date Reported")
plt.ylabel("Cumulative_deaths")
plt.title("Pakistan: Cumulative_deaths")
plt.show()
```

```python
def check_cum_deaths(country_name):
    country = who["Country"] == country_name
    df = who[country]
    plt.plot(df["Date_reported"], df["Cumulative_deaths"])
    plt.xlabel("Date_Reported")
    plt.ylabel("Cumulative_Deaths")
    plt.title("{} Cumulative deaths".format(country_name))
    plt.show()
    
```

```python
check_cum_deaths("India")
```

```python
check_cum_deaths("Poland")
```
