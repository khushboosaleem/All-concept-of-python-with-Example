```python
def col(col_name):
    col_name = col_name.lower()
    col_name = col_name.strip()
    col_name = col_name.replace("operating system","os")
    col_name = col_name.replace("(","").replace(")","").replace(" ","_")
    return col_name
    
```

```python
for i in laptop.columns:
    print(col(i))
```

```python
updt = []

for i in laptop.columns:
    updt.append(col(i))
```
