```python
html = pd.read_html("https://en.wikipedia.org/wiki/Economy_of_the_United_States")
html
```

```python
html[0]
```

```python
html[3]
```

```python
visual = html[3]
visual
```

```python
x = visual["Year"]
x
```

```python
y = visual["GDP growth (real)"]
y
```

```python
y = y.str.replace("%"," ").astype(float)
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(x, y, marker='o', linestyle='-', color='b', label='GDP Growth (Real)')

# Adding title and labels
plt.title('GDP Growth (Real) Over Years')
plt.xlabel('Year')
plt.ylabel('GDP Growth (Real) (%)')

# Adding a grid
plt.grid(True)

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```
